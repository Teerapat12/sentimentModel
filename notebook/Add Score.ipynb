```python
from collections import Counter
def computeJudgeScore(row):
    scores = [row['score_0'],row['score_1'],row['score_2']]

    if(row['scoredAmt']==1):
        return "1 Comment"
    elif(row['scoredAmt']==3):
        counts = Counter([row['score_0'],row['score_1'],row['score_2']])
        if(counts.most_common()[0][1]>1): #Found majority
            if(counts.most_common()[0][1]==3):
                return "Consensus of 3"
            return "2 out of 3"
        else: #Unique. 
            scores = [score if score>-32 else 0 for score in scores] #change -32 to 0
            scores.sort()
            if(scores[0]<0 and scores[1] < 0):#Negative Majority
                return "Negative Majority (3)"
            elif(scores[1]>0 and scores[2]>0):
                return "Positive Majority (3)"
            else:
                return "Neutral Majority (3)"
    elif(row['scoredAmt']==2):
        if(row['score_0']==row['score_1']):
            return "Consensus of 2"
        else:
            scores = [row['score_0'],row['score_1']]
            scores =  [score if score>-32 else 0 for score in scores] #change -32 to 0
            if(scores[0]<0 and scores[1]<0):
                return "Negative Majority (2)"
            elif(scores[0]>0 and scores[1]>0):
                return "Positive Majority (2)"
            elif(scores[0]==0 and scores[1]==0):
                return "Neutral Majority (2)"
            elif(abs(scores[0]-scores[1])<=1):
                return "Little Diff (2)"
            return "Diff Opinion (2)"
            #If two comment but with different score
df['judgeScoring'] = df.apply(computeJudgeScore,axis=1)
```

```python
from collections import Counter
def computeScore(row):
    scores = [row['score_0'],row['score_1'],row['score_2']]

    if(row['scoredAmt']==1):
        return row['score_0']
    elif(row['scoredAmt']==3):
        counts = Counter([row['score_0'],row['score_1'],row['score_2']])
        if(counts.most_common()[0][1]>1): #Found majority
            return counts.most_common()[0][0]
        else: #Unique. 
            scores = [score if score>-32 else 0 for score in scores] #change -32 to 0
            scores.sort()
            if(scores[0]<0 and scores[1] < 0):#Negative Majority
                return (scores[0]+scores[1])/2
            elif(scores[1]>0 and scores[2]>0):
                return (scores[1]+scores[2])/2
            else:
                return 0
    elif(row['scoredAmt']==2):
        if(row['score_0']==row['score_1']):
            return row['score_0']
        else:
            scores = [row['score_0'],row['score_1']]
            scores =  [score if score>-32 else 0 for score in scores] #change -32 to 0
            return (scores[0]+scores[1])/2
            #If two comment but with different score
df['score'] = df.apply(computeScore,axis=1)
```
